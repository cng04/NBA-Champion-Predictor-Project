## In this file, We Will Parse, Clean and Merge the Per Game and Advanced Stats in preparation for Machine Learning

### Parsing Per Game Stats

```python
from bs4 import BeautifulSoup
import pandas as pd
```

```python
with open("season_stats_2023.html") as f:
    html = f.read()
```

```python
# Parsing Per Game Stats
soup = BeautifulSoup(html, "html.parser")

# finds the per game stats table and returns an object representing the element
pg_table = soup.find(id="div_per_game-team")

# turn the per game stats table object into a pandas dataframe
pg_df = pd.read_html(str(pg_table))[0]

pg_df
```

```python
# Cleaning per_game stats

# Deleting some columns
del pg_df["Rk"]
del pg_df["G"]

# Removing the last row which is the "league average row"
pg_df = pg_df.drop(pg_df.index[-1])
pg_df
```

### Parsing Advanced Stats

```python
# Parsing Advanced Stats
soup = BeautifulSoup(html, "html.parser")

# find the class called "over_header"
over_header = soup.find(class_="over_header")

# remove the "over_header" class
over_header.decompose()

# find the table with the specified id and return an html object
adv_table = soup.find(id="div_advanced-team")

# turn the html object representing the advanced stats table into a pandas dataframe
adv_df = pd.read_html(str(adv_table))[0]

# drops the rows with NaN in the Rk column
adv_df = adv_df.dropna(subset="Rk")

# removes the columns that have "MOV" as a value in the MOV columns
adv_df = adv_df[adv_df["MOV"] != "MOV"]

# add the year 
adv_df["Year"] = 2023

adv_df
```

```python
# Cleaning advanced stats data
del adv_df["Unnamed: 17"]
del adv_df["Unnamed: 22"]
del adv_df["Unnamed: 27"]
del adv_df["Arena"]
del adv_df["Attend."]
del adv_df["Attend./G"]
del adv_df["Age"]
del adv_df["W"]
del adv_df["L"]
del adv_df["PW"]
del adv_df["PL"]
del adv_df["Rk"]

adv_df.columns = adv_df.columns.str.replace(".1", "_opp")
adv_df
```

### Merge Per Game and Advanced Stats

```python
adv_pg = pd.merge(pg_df, adv_df, on="Team")
adv_pg
```

```python
adv_pg["Team"] = adv_pg["Team"].str.replace("*", "", regex=False)
```

```python
# Creating a dictionary from the team_abbrev csv file where the keys are the team's full name
# and the values are the abbreviated name

team_abbrev = {}

with open("team_names.csv") as file:
    entries = file.readlines()
    for entry in entries[1:]:
        abbrev, full_name = entry.replace("\n", "").split(",")
        team_abbrev[full_name] = abbrev

adv_pg["Team"] = adv_pg["Team"].map(team_abbrev)
adv_pg
```

```python
from sklearn.preprocessing import MinMaxScaler

# scaling the data to values between 0 (smallest value) to 1 (largest value) using the MinMaxScaler Class from 
# sk.learnprocessing library

scaler = MinMaxScaler()

avoid_scale = ["Year", "Team", "Score"]

# choosing which columns to scale
scale_cols = adv_pg.columns[~adv_pg.columns.isin(avoid_scale)]

# scaling the chosen columns 
adv_pg[scale_cols] = scaler.fit_transform(adv_pg[scale_cols])

adv_pg
```

```python
adv_pg.to_csv("2023_team_stats.csv")
```
