# In This File, We Will Do Some More Data Cleaning and Merge the Per Game Stats, Playoff Scores and Advanced Stats

# Further Data Cleaning

```python
from bs4 import BeautifulSoup
import pandas as pd
import os
from sklearn.preprocessing import MinMaxScaler
```

```python
t_stats_df = pd.read_csv("CSV/team_stats.csv")
del t_stats_df["Unnamed: 0"]
```

```python
t_stats_df
```

**Combine teams that have different historical names**

```python
replace_dict = {"NJN": "BRK", "CHA": "CHO", "CHH": "CHO", "VAN": "MEM", "NOH": "NOP", "NOK": "NOP", "SEA": "OKC", "WSB": "WAS"}
t_stats_df["Team"] = t_stats_df["Team"].replace(replace_dict)
```

```python
t_stats_df
```

## Merging Team Stats, Playoff Scores and Advanced Stats Dataframes

```python
# make playoff scores dataframe
t_score_df = pd.read_csv("CSV/playoff_score.csv")
del t_score_df["Unnamed: 0"]
```

```python
t_score_df
```

```python
# make advanced stats_dataframe
adv_stats = pd.read_csv("CSV/advanced_stats.csv")
del adv_stats["Unnamed: 0"]
adv_stats
```

```python
# merge the team stats, playoff scores and advanced stats dataframe on the "Team" and "Year" columns
# how=left means a left join meaning all rows from the larger dataframe are kept, only matching rows from 
# the smaller dataframe are merged

combined_df = pd.merge(t_stats_df, t_score_df, on=["Team", "Year"], how="left").merge(adv_stats, on=["Team", "Year"], how="left")
```

```python
combined_df["Score"].fillna(0, inplace=True)
```

```python
combined_df
```

```python
# scaling the data to values between 0 (smallest value) to 1 (largest value) using the MinMaxScaler Class from 
# sk.learnprocessing library

scaler = MinMaxScaler()

avoid_scale = ["Year", "Team", "Score"]

# choosing which columns to scale
scale_cols = combined_df.columns[~combined_df.columns.isin(avoid_scale)]

# scaling the chosen columns 
combined_df[scale_cols] = scaler.fit_transform(combined_df[scale_cols])

combined_df
```

```python
combined_df = combined_df.sort_values(["Year", "Score"], ascending=[True, False])
```

```python
combined_df
```

```python
# getting a merged dataframe with per game stats, advanced stats and playoff scores
combined_df.to_csv("CSV/merged_stats_scores.csv")
```
